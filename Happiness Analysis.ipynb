#### Firstly, we created a new data frame named "df1" that contains the attributes related to the personal information in the survey. Then we split the data with 80% goes into training data and the rest goes into the test data.

```python
DataTree = GET_DATA_TREE(df1)
X_train, X_test, Y_train, Y_test = GET_DATA_FOR_TREE(DataTree,splitrate = 0.2)
```

#### We build three tree-based models with the training data, and they are decision tree model, random forest model, and GBM model. Then we made a prediction on the test dataset and evaluate each model based on the predictive accuracy.

```python
dtc,rfc,gbc,dtc_y_predict,rfc_y_predict,gbc_y_predict = BULID_DT(X_train, X_test, Y_train, Y_test)
```
